```python
%matplotlib inline

from pathlib import Path

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
import matplotlib.pylab as plt
from dmba import classificationSummary, gainsChart
```

```python
df = pd.read_csv("./sample.csv", encoding="CP949")
```

```python
df = df.drop(columns=['Unnamed: 0','Unnamed: 93'])
```

```python
df['Unnamed: 94'] = df['Unnamed: 94'].replace({'기타 변경':'기타변경','간섭':'기타변경'})
```

```python
for i in df.columns:
    df[i] = df[i].astype('category')
```

```python
df.info().head()
```

```python
predictor = df.drop(columns='Unnamed: 94')
outcome = 'Unnamed: 94'
classes = df['Unnamed: 94'].unique()

X = predictor
y = df[outcome]

X_train, X_vaild, y_train, y_vaild = train_test_split(X, y, test_size=0.2, random_state=1,stratify=y)
```

```python
y.value_counts()
```

```python
#모델 적합
class_nb = MultinomialNB(alpha=0.01)
class_nb.fit(X_train, y_train)

# 적합값
predProb_train = class_nb.predict_proba(X_train)
predProb_test = class_nb.predict_proba(X_vaild)

# 예측값
y_train_pred = class_nb.predict(X_train)
y_vaild_pred = class_nb.predict(X_vaild)
```

```python
classificationSummary(y_train, y_train_pred, class_names=classes)

print()

classificationSummary(y_vaild, y_vaild_pred, class_names=classes)
```

```python
prob_x = pd.DataFrame(predProb_X,columns = classes)
```
